## Univariate analysis 

```python
#reusable ploting function
def plotfunc(df,col,title,hue = None):
    
    sns.set_style('darkgrid')
    sns.set_context('poster')
    plt.rcParams["axes.labelsize"] = 15
    plt.rcParams['axes.titlesize'] = 18
    plt.rcParams['axes.titlepad'] = 20
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.yscale('log')
    plt.title(title,fontsize=14)
    ax = sns.countplot(data=df, x=col, hue=hue, palette='bright') 
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

    plt.show()
```

```python
# Function for box plot
def cusBoxPlot(data, col, title):
    plt.figure(figsize=(12, 8))
    plt.boxplot(data[col], vert=True, patch_artist=True, boxprops=dict(facecolor='orange'))
    plt.title(title, fontsize=18)
    plt.xlabel(col, fontsize=20)
    plt.ylabel('Income Amount', fontsize=16)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.show()

```
